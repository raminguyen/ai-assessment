```python
import json
import re

# File paths
FILE_PATH = "/Users/ramihuunguyen/Documents/PhD/AI-Assessment/Essay/Models/Automation/gemini_step1_draft1.json"
OUT_PATH = "essay_extracted.txt"

# Read JSON file
with open(FILE_PATH, 'r', encoding='utf-8') as f:
    data = json.load(f)

draft_text = data.get('draft_text', '')

# Extract content between <result> and </result>
pattern = r'<result>\\n(.*?)\\n</result>'
matches = re.findall(pattern, draft_text, re.DOTALL)

if matches:
    essay = matches[0]  # Take first match only
    
    # Clean escape sequences
    essay = essay.replace('\\n', '\n')
    
    essay = essay.replace('\\\\', '\\')
    essay = essay.replace("\\'", "'")
    essay = essay.replace('\\"', '"')
    essay = essay.replace('\\u2014', '—')
    essay = essay.replace('\\u2013', '–')
    essay = essay.replace('\\u2019', "'")
    essay = essay.replace('\\u00e1', 'á')
    
    # Save to file
    with open(OUT_PATH, 'w', encoding='utf-8') as f:
        f.write(essay)
    
    print(f"Essay extracted: {len(matches)} found, saved first one")
    print(f"Output: {OUT_PATH}")
    print(f"Words: ~{len(essay.split())}")
else:
    print("No essay found")
```
