```python
import json
import re

# File paths
FILE_PATH = "/Users/ramihuunguyen/Documents/PhD/AI-Assessment/Essay/Models/Automation/gemini_step1_draft1.json"
OUT_PATH = "essay_extracted.txt"

# Read JSON file
with open(FILE_PATH, 'r', encoding='utf-8') as f:
    data = json.load(f)

draft_text = data.get('draft_text', '')

# Extract content between <result> and </result>
pattern = r'<result>\\n(.*?)\\n</result>'
matches = re.findall(pattern, draft_text, re.DOTALL)

if matches:
    essay = matches[0]  # Take first match only
    
    # Clean escape sequences
    essay = essay.replace('\\n', '\n')
    
    essay = essay.replace('\\\\', '\\')
    essay = essay.replace("\\'", "'")
    essay = essay.replace('\\"', '"')
    essay = essay.replace('\\u2014', '—')
    essay = essay.replace('\\u2013', '–')
    essay = essay.replace('\\u2019', "'")
    essay = essay.replace('\\u00e1', 'á')
    
    # Save to file
    with open(OUT_PATH, 'w', encoding='utf-8') as f:
        f.write(essay)
    
    print(f"Essay extracted: {len(matches)} found, saved first one")
    print(f"Output: {OUT_PATH}")
    print(f"Words: ~{len(essay.split())}")
else:
    print("No essay found")
```

```python
# --- remove line breaks from essay_extracted.txt and save clean version ---
input_path = "essay_extracted.txt"
output_path = "essay_extracted.txt"

with open(input_path, "r", encoding="utf-8") as f:
    text = f.read()

# Remove all line breaks and extra spaces
clean_text = " ".join(text.split())

# Save cleaned text to new file
with open(output_path, "w", encoding="utf-8") as f:
    f.write(clean_text)

print("✅ Saved:", output_path)
```

```python
# --- PARAMS ---
JSON_PATH = "/Users/ramihuunguyen/Documents/PhD/AI-Assessment/Essay/Models/Automation/outputs/gemini_step2_draft1.json"
STRIP_MARKDOWN = True  # set False to keep original formatting

# --- RUN ---
import json, re, os, sys

def strip_md_html(s: str) -> str:
    s = re.sub(r"<[^>]+>", "", s)                 # remove HTML-ish tags
    s = re.sub(r"[*_`#>-]+", "", s)               # remove common markdown marks
    s = re.sub(r"\n{2,}", "\n", s)
    return s.strip()

def extract_result_blocks(text: str):
    return re.findall(r"<result>\s*(.*?)\s*</result>", text, flags=re.DOTALL)

def walk_collect_strings(x):
    """collect all string fields from arbitrary JSON (dict/list/nested)."""
    if isinstance(x, dict):
        for v in x.values():
            yield from walk_collect_strings(v)
    elif isinstance(x, list):
        for v in x:
            yield from walk_collect_strings(v)
    elif isinstance(x, str):
        yield x

# load JSON
with open(JSON_PATH, "r", encoding="utf-8") as f:
    data = json.load(f)

# search every string in the JSON for <result>...</result>
all_strings = list(walk_collect_strings(data))
found_blocks = []
for s in all_strings:
    blocks = extract_result_blocks(s)
    if blocks:
        found_blocks.extend(blocks)

if not found_blocks:
    print("⚠️ No <result>...</result> blocks found in the JSON.")
    sys.exit(0)

# print clean results
for i, block in enumerate(found_blocks, 1):
    out = strip_md_html(block) if STRIP_MARKDOWN else block
    print(f"\n===== RESULT {i} =====\n{out}\n")
```
